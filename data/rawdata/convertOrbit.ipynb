# Read in Mark Hammergren's orbit and convert it to a JSON file

*Here's an orbital integration going back 1000 days, a little more than one orbit, with smaller timesteps closer to the impact time.* 

*The positions are in cartesian heliocentric ecliptic coordinates referenced to ecliptic epoch J2000. The orbital elements are also referenced to J2000, with the epoch of osculation being each listed JD. 
There's some problem with my transformation from position and velocity to orbital elements for the mean anomaly, something to due with the trigonometric quadrant. Near in time to the impact, it should be OK. I'm not going to delay sending this to you while I figure that out.*

*The columns are:*

**JD, x, y, z, semimajor axis, eccentricity, inclination, longitude of the ascending node, argument of the perihelion, mean anomaly**

```python
import pandas as pd
```

```python
names = ['JD', 'x', 'y', 'z', 'a', 'e', 'i', 'la', 'ap', 'ma']
orbit = pd.read_csv('Wisconsin meteor day-1000.txt', delim_whitespace=True, header=None, names=names)
print(orbit)
```

### Write to JSON

```python
orbit.to_json('../HammergrenOrbit.json')
```
